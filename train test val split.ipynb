```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('https://www.dropbox.com/s/7pdx8pmoy7kk33y/scaled_airbnb_1201.csv?dl=1')
```

```python
df_o = pd.read_csv('https://www.dropbox.com/s/73tywzf3o95lggx/original_airbnb_1201.csv?dl=1')
```

```python
df_o['neigborhood_cat'] = df['neigborhood_cat']
```

```python
df['neighborhood_room'] = df['neigborhood_cat'].astype(str) + df['room_type'].astype(str)
```

```python
X = df[['neigborhood_cat', 'room_type',
       'minimum_nights', 'number_of_reviews', 'reviews_per_month',
       'calculated_host_listings_count', 'availability_365',
       'number_of_reviews_ltm', 'days_since_last_review']]
y = df['price']
```

```python
import random

random.seed(50)
```

```python
# test 20%, train 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    stratify=df['neighborhood_room'], 
                                                    test_size=0.40)
```

```python
X_train_o = df_o.iloc[X_train.index][['neigborhood_cat', 'room_type',
       'minimum_nights', 'number_of_reviews', 'reviews_per_month',
       'calculated_host_listings_count', 'availability_365',
       'number_of_reviews_ltm', 'days_since_last_review']]
y_train_o = df_o.iloc[X_train.index]['price']

X_test_o = df_o.iloc[X_test.index][['neigborhood_cat', 'room_type',
       'minimum_nights', 'number_of_reviews', 'reviews_per_month',
       'calculated_host_listings_count', 'availability_365',
       'number_of_reviews_ltm', 'days_since_last_review']]
y_test_o = df_o.iloc[X_test.index]['price']
```

```python
train  = pd.concat([X_train, y_train], axis=1)
train_o = pd.concat([X_train_o, y_train_o], axis=1)
test = pd.concat([X_test, y_test], axis=1)
test_o = pd.concat([X_test_o, y_test_o], axis=1)
```

```python
train.to_csv('train4.csv')
```

```python
train_o.to_csv('train_o4.csv')
```

```python
test.to_csv('test4.csv')
```

```python
test_o.to_csv('test_o4.csv')
```
