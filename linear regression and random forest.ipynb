```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("https://www.dropbox.com/s/09syh5fwp3ggp11/cleaned_airbnb.csv?dl=1")
```

```python
df.head()
```

```python
df.columns
```

```python
X_cat = ['neighbourhood', 'room_type']
```

```python
for i in X_cat:
  one_hot = pd.get_dummies(df[i], drop_first = True).rename(columns=lambda x: i + '_' + str(x))
  df = pd.concat([df, one_hot], axis=1)
  df.drop([i], inplace=True, axis=1)
```

```python
df.columns
```

```python
y = df['price']
X = df[['minimum_nights', 'number_of_reviews', 'reviews_per_month',
       'calculated_host_listings_count', 'availability_365',
       'number_of_reviews_ltm', 'days_since_last_review',
       'neighbourhood_Barnet', 'neighbourhood_Bexley', 'neighbourhood_Brent',
       'neighbourhood_Bromley', 'neighbourhood_Camden',
       'neighbourhood_City of London', 'neighbourhood_Croydon',
       'neighbourhood_Ealing', 'neighbourhood_Enfield',
       'neighbourhood_Greenwich', 'neighbourhood_Hackney',
       'neighbourhood_Hammersmith and Fulham', 'neighbourhood_Haringey',
       'neighbourhood_Harrow', 'neighbourhood_Havering',
       'neighbourhood_Hillingdon', 'neighbourhood_Hounslow',
       'neighbourhood_Islington', 'neighbourhood_Kensington and Chelsea',
       'neighbourhood_Kingston upon Thames', 'neighbourhood_Lambeth',
       'neighbourhood_Lewisham', 'neighbourhood_Merton',
       'neighbourhood_Newham', 'neighbourhood_Redbridge',
       'neighbourhood_Richmond upon Thames', 'neighbourhood_Southwark',
       'neighbourhood_Sutton', 'neighbourhood_Tower Hamlets',
       'neighbourhood_Waltham Forest', 'neighbourhood_Wandsworth',
       'neighbourhood_Westminster', 'room_type_Hotel room',
       'room_type_Private room', 'room_type_Shared room']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression().fit(X_train, y_train)
```

```python
lr.score(X_train,y_train), lr.score(X_test, y_test)
```

```python
import statsmodels.api as sm
```

```python
lr2 = sm.OLS(y_train, sm.add_constant(X_train))
```

```python
result = lr2.fit()
print(result.summary2())
```

```python
# random forest regressor 
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(max_depth=2, random_state=0)
rf.fit(X_train, y_train)
```

```python
rf.score(X_train, y_train), rf.score(X_test, y_test)
```
